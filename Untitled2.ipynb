```python
import mne
from pyedflib import highlevel
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Step 1: Reading the EEG data and annotations
# Replace 'your_path' with the actual path to your EDF files
eeg_file = 'SC4022E0-PSG.edf'
hypnogram_file = 'SC4022EJ-Hypnogram.edf'

raw = mne.io.read_raw_edf(eeg_file, preload=True)
annotations = mne.read_annotations(hypnogram_file)
raw.set_annotations(annotations, emit_warning=False)

# Map annotations to integers for sleep stages
annotation_desc_2_event_id = {
    'Sleep stage W': 1,
    'Sleep stage 1': 2,
    'Sleep stage 2': 3,
    'Sleep stage 3': 4,
    'Sleep stage 4': 4,
    'Sleep stage R': 5
}
events, _ = mne.events_from_annotations(
    raw,
    event_id=annotation_desc_2_event_id,
    chunk_duration=30.0,
)

# Step 2: Preprocess the data (filtering)
raw.filter(0.5, 30., fir_design='firwin')

# Step 3: Epoching
tmax = 30. - 1. / raw.info['sfreq']  # tmax in seconds
epochs = mne.Epochs(raw=raw, events=events, event_id=annotation_desc_2_event_id,
                    tmin=0., tmax=tmax, baseline=None, preload=True)

import mne
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

# Assuming you have already loaded the EEG data into `epochs`

# Step 4: Feature Extraction (Power Spectral Density using Welch's method)
# Use psd_welch instead of psd_multitaper
psds, freqs = mne.time_frequency.psd_welch(epochs, fmin=0.5, fmax=30, n_fft=256, n_per_seg=128)
psds = 10 * np.log10(psds)  # Convert power to dB scale
psds_mean = psds.mean(axis=-1)  # Mean power spectral density across frequencies

# The rest of the steps (5 to 8) remain the same
labels = epochs.events[:, -1]
X_train, X_test, y_train, y_test = train_test_split(psds_mean, labels, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from scipy.signal import periodogram
import numpy as np

def compute_mean_psd(epochs_data, sfreq):
    """
    Compute mean PSD for each epoch using periodogram.
    
    Parameters:
    - epochs_data: array of shape (n_epochs, n_channels, n_times)
    - sfreq: Sampling frequency
    
    Returns:
    - mean_psds: Mean PSD of each epoch
    """
    n_epochs, n_channels, n_times = epochs_data.shape
    psds = np.empty((n_epochs, n_channels, n_times // 2 + 1))
    
    for i in range(n_epochs):
        for j in range(n_channels):
            f, Pxx = periodogram(epochs_data[i, j, :], sfreq, scaling='spectrum')
            psds[i, j, :] = Pxx
    
    mean_psds = psds.mean(axis=1)  # Mean across channels
    return mean_psds

# Extract the data from the MNE Epochs object
epochs_data = epochs.get_data()
sfreq = epochs.info['sfreq']

# Compute the mean PSD
mean_psds = compute_mean_psd(epochs_data, sfreq)

# Continue with your feature selection, model training, etc., using mean_psds as features
```
